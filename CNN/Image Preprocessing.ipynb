```python
from PIL import Image
import os

def resize_images_in_folder(folder_path, output_folder, size=(256, 256)):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        output_path = os.path.join(output_folder, image_file)

        try:
            with Image.open(image_path) as img:
                # Resize image to 256x256 pixels
                resized_img = img.resize(size, Image.ANTIALIAS)
                # Save the resized image
                resized_img.save(output_path)

        except Exception as e:
            print(f"Error processing image {image_path}: {e}")

def resize_images_in_subfolders(root_folder, output_root_folder, size=(256, 256)):
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        output_folder = os.path.join(output_root_folder, folder_name)

        if os.path.isdir(folder_path):
            print(f"\nResizing images in folder: {folder_path}")
            resize_images_in_folder(folder_path, output_folder, size)

if __name__ == "__main__":
    root_directory = "C:/Users/Ahmad/Desktop/Signals Project 2/Dataset/Training"  # Replace with the path to your root directory
    output_directory = "C:/Users/Ahmad/Desktop/Signals Project 2/Dataset/Training2"  # Replace with the path to the output directory
    target_size = (256, 256)

    resize_images_in_subfolders(root_directory, output_directory, target_size)
    print("\nResizing completed.")
```

```python
from PIL import Image
import os

def print_grayscale_images(folder_path):
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    
    grayscale_images = []

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)

        try:
            with Image.open(image_path) as img:
                if img.mode == 'L':
                    grayscale_images.append(image_path)
        except Exception as e:
            print(f"Error processing image {image_path}: {e}")

    if grayscale_images:
        print(f"\nGrayscale images in folder {folder_path}:")
        for img_path in grayscale_images:
            print(img_path)
    else:
        print(f"\nNo grayscale images found in folder {folder_path}.")

# Example usage:
# Replace "/path/to/your/images" with the actual path to your images directory
folder_path = "C:/Users/Ahmad/Desktop/Signals Project 2/Dataset/Training2/meningioma"
print_grayscale_images(folder_path)
```
